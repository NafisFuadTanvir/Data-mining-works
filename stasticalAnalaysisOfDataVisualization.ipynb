```python
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/titanic.csv")

data.head()
```

```python
sns.displot(data['Age'],bins=10,kde=True,rug=False)
```

```python
sns.displot(data['Embarked'],bins=10,kde=True,rug=False)
```

```python
data["Catagorical_Age"]=pd.cut(x=data['Age'],bins=[0,3,17,63,99],labels=["Baby","Child","Adult","Elderly"])
```

```python
sns.catplot(x="Catagorical_Age",y="Age",kind="box",data=data)
```

```python
sns.violinplot(x="Catagorical_Age",y="Age",data=data,size=6)
```

```python
test_data=pd.DataFrame(dict(time=pd.date_range("2017-1-1",periods=500),value=np.random.randn(500).cumsum()))
sns.relplot(x="time",y="value",kind="line",data=test_data,height=5, aspect=15/6)
```

```python
sns.catplot(x="Sex",y="Survived",kind="bar",data=data)
```

```python
#scatte-plot
sns.jointplot(x="Age",y="Fare",kind="scatter",data=data)
```
